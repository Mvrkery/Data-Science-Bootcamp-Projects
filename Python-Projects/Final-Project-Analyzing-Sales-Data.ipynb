```python
# TODO 04 - your friend ask for all order data in `California` and `Texas` in 2017 (look at Order Date), send him csv file
cal_tex_2017 = df[(df['State'].isin(['California', 'Texas'])) & (df['Order Date'].dt.year == 2017)]
cal_tex_2017.to_csv('California_Texas_2017.csv')
cal_tex_2017
```

```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
df[df['Order Date'].dt.year == 2017]['Sales'].agg(['sum','mean','std'])
```

```python
# TODO 06 - which Segment has the highest profit in 2018
df[df['Order Date'].dt.year == 2018 ].groupby('Segment')['Profit'].sum()
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
df[ (df['Order Date'] >= '2019-04-15 00:00:00') & (df['Order Date'] <= '2019-12-31 23:59:59')].groupby('State')['Sales'].sum().sort_values().head(5)
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
W_CT_sales_2019 = df[((df['Region'] == 'West') | (df['Region'] == 'Central')) & (df['Order Date'].dt.year == 2019)]['Sales'].sum()
total_sales_2019 = df[df['Order Date'].dt.year == 2019]['Sales'].sum()
percentage = W_CT_sales_2019/total_sales_2019 * 100
percentage

print(f"The total sales in 2019 is ${total_sales_2019:,.2f}.\
      \nThe total sales in the West and Central in 2019 is ${W_CT_sales_2019:,.2f}.\
        \nThe proportion of total sales (%) in West + Central in 2019 is {percentage:.2f}%")
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
date19to20 = df[(df['Order Date'].dt.year >= 2019) & (df['Order Date'].dt.year <= 2020)]
top10_Sales = date19to20.groupby("Product Name")['Sales'].sum().sort_values(ascending=False).reset_index().head(10)

top10_order = date19to20.groupby("Product Name")['Order ID'].count().sort_values(ascending=False).reset_index().head(10)

print(top10_Sales)
print(top10_order)
merged_top10 = pd.merge(top10_Sales, top10_order, on='Product Name')
merged_top10
top10_Sales
top10_order
```
