```r
library(rvest)
library(tidyverse)

url <- "https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc"

movie_name <- url %>%
  read_html() %>%
  html_elements("h3.lister-item-header") %>%
  html_text2()

ratings <- url %>%
  read_html() %>%
  html_elements("div.ratings-imdb-rating") %>%
  html_text2() %>%
  as.numeric()

votes <- url %>%
  read_html() %>%
  html_elements("p.sort-num_votes-visible") %>%
  html_text2()

df <- data.frame(movie_name, ratings, votes)

df <- df %>%
  separate(votes, sep="\\|", into=c("votes", "gross", "tops")) %>%
  mutate(
    tops = ifelse(is.na(tops), gross, tops),
    gross = ifelse(grepl("Top 250: #", gross), "Coming Soon", gross)
  )

View(df)
```
